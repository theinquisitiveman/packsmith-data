```python
import folium
import pandas as pd
from folium.plugins import HeatMap
```

```python
# The data I'm pulling is for all of 2024

df = pd.read_csv('/Users/theinquisitiveman/Documents/fulfillment_data.csv')

display(df.head())
```

```python
# Pulling destination locations for delviered items to the United States
m = folium.Map(location=[df['to_lat'].mean(), df['to_lon'].mean()], zoom_start=3)
```

```python
heat_data = [[row['to_lat'], row['to_lon']] for index, row in df.iterrows()]

HeatMap(heat_data).add_to(m)
```

```python
# This is gonna group 'to_province' and count the entries, but really it's for the state
state_counts = df.groupby('to_province').size().reset_index(name='count')

# Calculating the percentage of orders it represents
total_orders = state_counts['count'].sum()
state_counts['percentage'] = (state_counts['count'] / total_orders) * 100
state_counts['percentage'] = state_counts['percentage'].apply(lambda x: f"{x:.2f}%")

# Display results
display(state_counts)
```
